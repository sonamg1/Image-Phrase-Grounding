# Preprocess VQA 2.0

```python
import json

from collections import Counter
```

```python
# this path refer to the file download from VQA2.0 website
train_anno_path = "/Users/david/desktop/CV_research/VQA/Annotations/v2_mscoco_train2014_annotations.json"
val_anno_path = "/Users/david/desktop/CV_research/VQA/Annotations/v2_mscoco_val2014_annotations.json"

train_ques_path = "/Users/david/desktop/CV_research/VQA/Questions/v2_OpenEnded_mscoco_train2014_questions.json"
val_ques_path = "/Users/david/desktop/CV_research/VQA/Questions/v2_OpenEnded_mscoco_val2014_questions.json"
testdev_ques_path = "/Users/david/desktop/CV_research/VQA/Questions/v2_Questions_Test_mscoco/v2_OpenEnded_mscoco_test-dev2015_questions.json"
test_ques_path = "/Users/david/desktop/CV_research/VQA/Questions/v2_Questions_Test_mscoco/v2_OpenEnded_mscoco_test2015_questions.json"
```

```python
def num_to_score(num):
    if num > 3:
        return 1
    elif num == 3:
        return 0.9
    elif num == 2:
        return 0.6
    elif num == 1:
        return 0.3
    elif num == 0:
        return 0
    else:
        raise TypeError("Wrong type of number!")
```

```python
def save_dataset(subtype, ques_path, anno_path=None):
    dataset = []
    imdir = "/Users/david/desktop/CV_research/VQA/Images/%s/COCO_%s_%012d.jpg"
    ques_data = json.load(open(ques_path, "r"))
    anno_data = json.load(open(anno_path, "r")) if anno_path is not None else None
    
    for i in range(len(ques_data["questions"])):
        ques = ques_data["questions"][i]["question"]
        ques_id = ques_data["questions"][i]["question_id"]
        img_id = ques_data["questions"][i]["image_id"]
        image_path = imdir%(subtype, subtype, img_id)
        
        item = {"ques_id": [ques_id], "img_path": image_path, "ques": [ques], "id": img_id}
        
        if anno_path is not None:
            mc_ans = anno_data["annotations"][i]["multiple_choice_answer"]
            answers = Counter()
            for ans in anno_data["annotations"][i]["answers"]:
                answers.update([ans["answer"]])
            answers = [(ans, num_to_score(num)) for ans, num in answers.items()]
            
            assert img_id == anno_data["annotations"][i]["image_id"], "Image index doesn't match!"
            assert ques_id == anno_data["annotations"][i]["question_id"], "Question index doesn't match!"
            
            item["mc_ans"] = [mc_ans]
            item["ans"] = [answers]
        dataset.append(item)
        if (i+1) % 1000 == 0:
            print("processing %i/%i" % (i, len(ques_data["questions"])))
        
    return dataset
```

```python
trainset = save_dataset("train2014", train_ques_path, train_anno_path)
```
